```python
import os
from matplotlib import pyplot as plt
```

```python
def direrctory_finder(path):
    directory = []
    for dir in os.listdir(path):
        if os.path.isdir(path+dir):
            directory.append(dir)
    return directory
```

```python
def dataset_counter(path):
    directories = direrctory_finder(path)
    count = {}
    for dir in directories:
        dirPath = path + dir + '/'
        count[dir] = len(os.listdir(dirPath))
    return count
```

```python
def show_dataset_count(path):
    data_count = dataset_counter(path)
    x = data_count.keys()
    y = data_count.values()

    title = path[path[:len(path)-1].rfind('/')+1:len(path)-1]
    
    fig, ax = plt.subplots()
    graph = ax.bar(x,y)
    ax.set(ylabel='count', xlabel = 'class', title=title)
    ax.bar_label(graph, fmt='{:,.0f}')
    plt.show()
```
